```python
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt #for plotting
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.metrics import mean_squared_error

X = dataset[['Years']]
y = dataset['CO2 Level (ppm)']

#X, y = dataset(n_samples=100, n_features=1, noise=20, random_state=0) #noise = bias and std. dev of the guassian noise, #random set for same data points for each run
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, train_size=0.75, random_state=0)

regressor = LinearRegression()
regressor.fit(X_train, y_train)
predictions = regressor.predict(X_test)
    
mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)

y_pred_line = regressor.predict(X)
new_X = [[2020]]
print(regressor.predict(new_X))
cmap = plt.get_cmap('viridis')
fig = plt.figure(figsize=(8,6))
m1 = plt.scatter(X_train, y_train, color=cmap(0.9), s=10)
m2 = plt.scatter(X_test, y_test, color=cmap(0.5), s=10)
plt.plot(X, y_pred_line, color='black', linewidth=2, label="Prediction")
plt.show()
```
