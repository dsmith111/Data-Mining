```python
import matplotlib.pyplot as plt
import numpy as np
import random
from __future__ import division
import math

def find_y(y_array, x_array, x):
    # y = mx + b
    m = (y_array[1] - y_array[0])/(x_array[1] - x_array[0])
    b = y_array[1] - (m*x_array[1])
    
    y = (m*x) + b
    return y


# Method which uses the random library to return a weight value between -1 and 1.
def initialize_weights(order):
    order += 1
    # Initialize Weights
    w = []

    for i in range(order):
        w.append(random.random()*2 - 1)
        
    return w

# PLA
def PLA(w, x, y, learning_rate, status):
    temp_w = np.array(w).T
    A = np.array([1, x, y])

    guess = int(temp_w.dot(A) >= 0)
    if guess == status:
        return w
    else:
        learning_array = A*learning_rate
        
        if (guess == 1):
            w = w - learning_array
            
        else:
            w = w + learning_array
    return w
            
# In-sample error
def sample_error(x, y, w, status, size):
    temp_w = np.array(w).T
    error = 0
    for i in range(size):
        A = np.array([1, x[i], y[i]])
        guess = int(temp_w.dot(A) >= 0)
        if guess != status[i]:
            error += 1
    in_sample_error = error/size
    return in_sample_error


# Define a rectangular box in the XY plane with coordinates (x1, y1) and (x2, y2) as (-2,-2)/(2,2)

boundaryx = [-2, 2]
boundaryy = [-2, 2]
size = 1000

plt.figure(0)

plt.scatter(boundaryx, boundaryy, c = 'white', marker ='.')

# Draw a random line across the box
line_x = [-2, 2]
line_y = [random.randint(-2, 2), random.randint(-2, 2)]
plt.plot(line_x, line_y, '-k', label = "Separation Line")

#plot 1000 points randomly in the box labeled 1 if above the line and...
# 0 if below

point_loc = [np.random.rand(size)*4 - 2, np.random.rand(size)*4 - 2] 
points = {'x':point_loc[0],'y':point_loc[1], 'status':np.zeros(size)}


for i in range(size):
    
    temp_y = find_y(line_y, line_x, points['x'][i])
    if points['y'][i] >= temp_y:
        points['status'][i] = 1
    else:
        points['status'][i] = 0


for i in range(size):
    if points['status'][i] == 1:
        h1 = plt.scatter(points['x'][i], points['y'][i], c = 'b')
    else:
        h2 = plt.scatter(points['x'][i], points['y'][i], c = 'r')
        

plt.legend((h1,h2),('"On" Data Points', '"Off" Data Points'))
plt.title("Initial Random Plot")
plt.show()


# Remove the line, then starting from random weights, use PLA to separate the data
plt.figure(1)

for i in range(size):
    if points['status'][i] == 1:
        h1 = plt.scatter(points['x'][i], points['y'][i], c = 'b')
    else:
        h2 = plt.scatter(points['x'][i], points['y'][i], c = 'r')

learning_rate = .01
w = initialize_weights(2)

print("_____________________________________________")


# Run PLA
iterations = 25
error = float('inf')
error_array = []
learning_change = []
for i in range(iterations):
    
    learning_change.append(learning_rate)
    for k in range(size):
        
        temp_w = PLA(w, points['x'][k], points['y'][k], learning_rate, points['status'][k])
        temp_error = sample_error(points['x'], points['y'], temp_w, points['status'], size)
        
        if temp_error <= error:
            error_array.append(temp_error)
            error = temp_error
            w = temp_w
            #learning_rate *= .90
    
    learning_rate *= .90
    

f = (-w[1]/w[2])*points['x'] - (w[0]/w[2])

plt.plot(points['x'], f, c = 'black')
plt.title("Perceptron Separated Graph")
plt.legend((h1,h2),('"On" Data Points', '"Off" Data Points'))
plt.figure(2)

plt.scatter(range(len(error_array)), error_array, c = 'b')
plt.title("In-sample Error")
plt.xlabel("Iteration")
plt.ylabel("Percent Error")

plt.figure(3)

plt.scatter(range(len(learning_change)), learning_change, c = 'g')
plt.title("Variation in Learning Rate")
plt.xlabel("Cycle")
plt.ylabel("Learning Rate")

plt.show()
print(w)
print(error)


```
